# Tính khoảng cách giữa 2 điểm

```python
# Tính khoảng cách giữa 2 điểm
import geopandas as gpd
from shapely.geometry import Point

# Tạo 2 điểm
point1 = Point(-122.224167, 37.176389)
point2 = Point(-122.419416, 37.774929)

# Tỉnh khoảng cách giữa 2 điểm
distance = point1.distance(point2)

# Show kết quả
print(distance)
```

# Tạo buffer từ 1 điểm

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon

# Tạo điểm
point = Point(10.775592424689332, 106.70302168729125)

# Tạo vùng đệm xung quanh điểm với bán kính 1000 mét
buffer = point.buffer(0.001)

# Tạo GeoDataFrame từ điểm và vùng đệm
geometry = [point, buffer]
gdf = gpd.GeoDataFrame(geometry=geometry, columns=['geometry'])

# Hiển thị điểm và vùng đệm trên bản đồ
gdf.plot(color=['yellow', 'green'])
```

# Tìm các điểm nằm trong bán kính cho phép

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon

# Tạo điểm
point = Point(10.775592424689332, 106.70302168729125)

# Tạo vùng đệm xung quanh điểm với bán kính 1000 mét

buffer = point.buffer(0.001)

# Tạo điểm từ tọa độ
point1 = Point(10.775619761168818, 106.70282689719542)
point2 = Point(10.776467190801428, 106.70175555166836)
point3 = Point(10.777751998919879, 106.70292429224334)

# Kiểm tra xem các điểm có nằm trong vùng đệm không
is_inside1 = buffer.contains(point1)
is_inside2 = buffer.contains(point2)
is_inside3 = buffer.contains(point3)

# Hiển thị vị trí các điểm so với vùng đệm
print(f"Point 1 is inside the buffer: {is_inside1}")
print(f"Point 2 is inside the buffer: {is_inside2}")
print(f"Point 3 is inside the buffer: {is_inside3}")

# Vẽ các điểm và vùng đệm trên bản đồ
geometry = [point, point1, point2, point3, buffer]
gdf = gpd.GeoDataFrame(geometry=geometry, columns=['geometry'])

gdf.plot(color=['yellow', 'red', 'red', 'red', 'green'])
```
