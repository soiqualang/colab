```
import requests
from requests.auth import HTTPDigestAuth
import json
```

```
#https://api.darksky.net/forecast/your-key/10.0180985638559,106.344112857092?lang=en&units=si
key='your-key'
lonlat='10.0180985638559,106.344112857092'
url='https://api.darksky.net/forecast/'+key+'/'+lonlat+'?lang=en&units=si'
print(url)
```

```
#myResponse = requests.get(url,auth=HTTPDigestAuth(raw_input("username: "), raw_input("Password: ")), verify=True)
myResponse = requests.get(url)
```

```
if(myResponse.ok):
    jData = json.loads(myResponse.content)
    print(jData);
    print(format(len(jData)))
    windSpeed=jData['currently']['windSpeed']
    windGust=jData['currently']['windGust']
    windBearing=jData['currently']['windBearing']
    print('windSpeed:',windSpeed)
    print('windGust:',windGust)
    print('windBearing:',windBearing)
else:
    myResponse.raise_for_status()
```
